```python
video_id = "AOwyd6l8KvM"

def make_ratios(video_id):
    subdf = subset.loc[subset.video_id == video_id]
    subdf['views_evo'] = round(subdf['view_count'].pct_change(1),3)
    subdf['likes_evo'] = round(subdf['likes'].pct_change(1),3)
    subdf['dislikes_evo'] = round(subdf['dislikes'].pct_change(1),3)
    subdf['comments_evo'] = round(subdf['comment_count'].pct_change(1),3)

    subdf = subdf.replace([np.inf, -np.inf], 0)

    subdf['trending_j'] = 1
    last = subdf.iloc[-1:]
    last['trending_j'] = 0
    subdf.update(last)
    subset.update(subdf)

    return subdf
```

```python
import tensorflow as tf
tf.random.set_seed(42)


model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(
    loss=tf.keras.losses.binary_crossentropy,
    optimizer=tf.keras.optimizers.Adam(lr=0.03),
    metrics=[
        tf.keras.metrics.BinaryAccuracy(name='accuracy'),
        tf.keras.metrics.Precision(name='precision'),
        tf.keras.metrics.Recall(name='recall')
    ]
)

history = model.fit(X_trainn, y_trainn, epochs=30)
```
